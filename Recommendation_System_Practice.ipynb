**we will find out content base by using cosain similarity**
the value of cosain simplirity lies between 0 and 1
zero mean no similarity and 1 mean similarity

**formula**

image.png

```python
from math import*
def seq_root(x):
 return round(sqrt(sum([a*a for a in x ])),3)

def cosine_similarity(x,y):
  numator = sum(a*b for a,b in zip(x,y))
  denumator = seq_root(x)*seq_root(y)
  return round(numator/float(denumator),3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# ----- Step 1: Load and Clean the Dataset -----
df = pd.read_csv('movies_metadata.csv')

# Ensure titles are strings and remove extra spaces
df['title'] = df['title'].astype(str).str.strip()
df['overview'] = df['overview'].fillna('')

# We'll use the 'overview' column as the content feature.
df['combined_features'] = df['overview']

# ----- Step 2: Vectorize the Text using TF-IDF -----
tfidf_vectorizer = TfidfVectorizer(
    stop_words='english',  # Remove common English words
    max_features=5000,     # Limit to top 5000 features for efficiency
    min_df=5               # Ignore terms that appear in fewer than 5 documents
                                    )
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined_features'])
print("TF-IDF matrix shape:", tfidf_matrix.shape)

# ----- Step 3: Use NearestNeighbors for Recommendations -----
# We set n_neighbors=11 to get 10 recommendations (the first neighbor is the movie itself)
nbrs = NearestNeighbors(n_neighbors=11, metric='cosine', algorithm='brute')
nbrs.fit(tfidf_matrix)

# Create a mapping from movie title to index
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

def get_recommendations(title):
    """
    Given a movie title, returns the top 10 recommended movies based on content similarity.
    Uses NearestNeighbors to avoid computing a full similarity matrix.
    """
    if title not in indices:
        print(f"Movie '{title}' not found. Please use one of these sample titles:")
        print(df['title'].dropna().head(10))
        return None

    idx = indices[title]
    # Find the nearest neighbors for the given movie's vector
    distances, indices_array = nbrs.kneighbors(tfidf_matrix[idx], n_neighbors=11)

    # Flatten the indices array and remove the first index (the movie itself)
    similar_indices = indices_array.flatten()[1:]

    # Retrieve the recommended movies from the DataFrame
    return df[['title', 'overview']].iloc[similar_indices]

```

```python
# ----- Step 4: Test the Recommendation Function -----
movie_title = input("Enter the movie title: ")  # Ensure this title exists exactly as shown in the dataset
recommended_movies = get_recommendations(movie_title)

if recommended_movies is not None and not recommended_movies.empty:
    print("\nTop recommended movies based on content similarity:")
    print(recommended_movies)
else:
    print("\nNo recommendations found. Please verify the movie title and dataset content.")
```

```python
# prompt: remove outlier of full dataset

import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler

# Assuming 'bank_en' DataFrame exists from the previous code
# Select numerical features for scaling (exclude non-numerical ones)
numerical_cols = bank_en1.select_dtypes(include=np.number).columns

# Initialize the RobustScaler
scaler = RobustScaler()

# Fit and transform the numerical features
bank_en1[numerical_cols] = scaler.fit_transform(bank_en1[numerical_cols])

# Now bank_en contains the scaled numerical features

#Remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Example usage for all numerical columns:
for col in numerical_cols:
    bank_en1 = remove_outliers_iqr(bank_en1, col)


```
